```python
import pandas as pd
import json 
import base64
import requests
```

```python
keys = pd.read_csv("spotifyAuth.txt", header = 0, sep = ",")
```

```python
appid = keys['Client_ID'][0]
```

```python
appsecret = keys['Client_Secret'][0]
```

```python
client_cred = base64.b64encode(str(appid + ":" + appsecret).encode('ascii'))
```

```python
header = {'Authorization': 'Basic {}'.format(client_cred.decode('ascii'))}
```

```python
payload = {'grant_type' : 'client_credentials'}
access_token_url = "https://accounts.spotify.com/api/token"
```

```python
response = requests.post(access_token_url, headers = header, data = payload)
```

```python
session_token = response.json()['access_token']
```

```python
session_header = {'Authorization': 'Bearer {}'.format(session_token)}
```

```python
midnights_id = "4X5cguwixSn1TppGbxdGRG"
```

```python
album_tracks_endpoint = "https://api.spotify.com/v1/albums/{}/tracks".format(midnights_id)
```

```python
midnights_tracks = requests.get(album_tracks_endpoint, headers = session_header)
```

```python
midnights_tracks.status_code
```

```python
midnights_dict = json.loads(midnights_tracks.text)
```
