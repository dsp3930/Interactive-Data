### #3 Set as category for sex/pid3/nonwhite/Evaluation

```python
csv_clean = csv.astype({'sex': 'category', 'pid3': 'category', 'nonwhite': 'category', 'Evaluation': 'category'})
```

 ### #5 one axis with a bar plot of the `Evaluation` variable
 

```python
fig2, ax2 = plt.subplots()
# Bar [center of bar,height]
dat = csv_clean.groupby('Evaluation').count()['caseid']
ax2.bar([0,1,2,3],dat)


plt.show()
```

### #6 One axis with a scatter plot of the `ConMediaUse` and `ImmiPhobia` variable

```python
fig, ax3 = plt.subplots()
ax3.scatter(csv_clean['ConMediaUse'], csv_clean['ImmiPhobia'], alpha = .05)
plt.show()
```
